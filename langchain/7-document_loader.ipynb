### Carregando Vídeos do Youtube

```python
from langchain_community.document_loaders.generic import GenericLoader
from langchain_community.document_loaders.blob_loaders.youtube_audio import YoutubeAudioLoader
from langchain.document_loaders.parsers import OpenAIWhisperParser
```

```python
url = "https://www.youtube.com/watch?v=6MECPST996I&t=456s"
save_dir = "files/youtube"
loader = GenericLoader(
    YoutubeAudioLoader([url],save_dir),
    OpenAIWhisperParser()
)
docs = loader.load()
```

```python
len(docs)
```

```python
docs[0].page_content[:200]
```

```python
pergunta = "Faça um resumo breve desse video para mim"
chain.run(input_documents=docs, question=pergunta)
```
