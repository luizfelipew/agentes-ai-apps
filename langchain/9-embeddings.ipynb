```python
from langchain_openai import OpenAIEmbeddings

embeddings_model = OpenAIEmbeddings()
```

```python
embeddings = embeddings_model.embed_documents(
    [
        "Eu gosto de cachorro",
        "Eu gosto de animais",
        "Hoje a tarde está chovendo"
    ]
)
```

```python
len(embeddings)
```

```python
embeddings[0][:10]
```

```python
len(embeddings[0])
```

```python
import numpy as np
# verificar a maior proximidade semântica entre os embeddings
np.dot(embeddings[0], embeddings[1])
```

```python
np.dot(embeddings[0], embeddings[2])
```

```python
np.dot(embeddings[1], embeddings[2])
```

```python
for i in range(len(embeddings)):
    for j in range(len(embeddings)):
        print(round(np.dot(embeddings[i], embeddings[j]), 2), end= " | ")
    print()
```

```python
pergunta = "O que é um cachorro?"

emb_query = embeddings_model.embed_query(pergunta)
emb_query[:10]
```
