```python
from dotenv import load_dotenv
from crewai import Crew, Process, Agent, Task

_ = load_dotenv()
```

```python
planejador_de_viagem = Agent(
    role="Planejador de viagem",
    goal="Planejar todos os detalhes de uma viagem, incluindo roteiros e atividades",
    backstory="""
    Você é uma especialista em planejamento de viagens, sempre em busca de novas
    aventuras e experiências incríveis. Seu objetivo é garantir os detlhes da viagem
    sejam organizados  de maneira eficiente e agradável.
    """,
    verbose=True
)
```

```python
orcamentista = Agent(
    role="Orçamentista de viagem",
    goal="Calcular o custo total da viagem considerando transporte, hospedagem, alimentação e atividades",
    backstory="""
    Você é um analista focado em viagem. Sua missão é garantir que os custos estejam
    dentro do orçamento, criando estimativas precisas para cada parte da viagem.
    """
)
```

```python
planeja_roteiro = Task(
    description="Crie um roteiro detalhado para a viagem na Europa, incluindo cidades, atividades e transporte",
    agent=planejador_de_viagem,
    expected_output=""""
    Um roteiro com a sequência  de cidades a serem visitadas, as principais 
    atividades e o tipo de transporte utilizado.
    """
)
```

```python
estima_orçamento = Task(
    description="""
    Calcule o orçamento total da viagem, levando em consideração 
    as cidades, transporte, hospedagem e atividades.
    """,
    agent=orcamentista,
    expected_output="""
    Uma estimativa de orçamento detalhada com custos aproximados para cada item da 
    viagem
    """
)
```

```python
viagem_crew = Crew(
    agents=[planejador_de_viagem, orcamentista],
    tasks=[planeja_roteiro, estima_orçamento],
    process=Process.sequential
    # verbose=True
)
```

```python
result = viagem_crew.kickoff()
print(result)
```
