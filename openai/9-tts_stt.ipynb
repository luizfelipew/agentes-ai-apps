## TTS


```python
import openai
import os
from dotenv import load_dotenv

load_dotenv()

client = openai.Client()
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
texto = """
Python é uma linguagem de programação popular e de fácil aprendizagem, conhecida por sua sintaxe clara e legibilidade. Ela é amplamente utilizada em diversas áreas, como desenvolvimento web, análise de dados, inteligência artificial, automação e muito mais. Com uma vasta biblioteca padrão e uma comunidade ativa, Python oferece ferramentas poderosas para desenvolvedores de todos os níveis.
Ela é amplamente utilizada em diversas áreas, como desenvolvimento web, análise de dados, inteligência artificial, automação e muito mais.
Principais características do python é sua versatilidade, permitindo se seja usado para projetos pequenos e grandes. 
Além disso python possui uma grande comunidade com uma vasta gama de bibliotecas e frameworks que facilitam o desenvolvimento de aplicações complexas.
Com uma ampla gama de tutoriais e recursos disponíveis, Python é uma escolha popular tanto para iniciantes quanto para desenvolvedores experientes.
"""
```

```python
resposta = client.audio.speech.create(
    model = "tts-1",
    voice = "alloy",
    input = texto
)
```

```python
arquivo = "files/voz.mp3"
resposta.write_to_file(arquivo)
```

## STT

```python
audio = open("files/voz.mp3", "rb")
transcript = client.audio.transcriptions.create(
    model="whisper-1",
    file=audio
)
```

```python
print(transcript.text)
```
